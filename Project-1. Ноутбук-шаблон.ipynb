1. Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:
* Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

*Совет: проверяйте полученные категории, например, с помощью метода unique()*


```python
#ваш код здесь
def hoice_educaton(education: str) -> str:
    """ Определяет тип образования по подовым словам в переданном тексте

    Args:
        education (str): Описание образования

    Returns:
        str: Тип образования, возможны вариынты: 'среднее специальное', 'среднее', 'неоконченное высшее', 'высшее'
    """    
    if education.lower().split()[:2] == ['среднее', 'специальное']:
        return 'среднее специальное'
    if education.lower().split()[0] == 'среднее':
        return 'среднее'
    if education.lower().split()[:2] == ['неоконченное', 'высшее']:
        return 'неоконченное высшее'
    if education.lower().split()[0] == 'высшее':
        return 'высшее'
    return 'other' # категория для проверкаи, появление ее в данных покажет ошибку в алгоритме
    
    
# Создаю новый признак
hh_df['Образование'] = hh_df['Образование и ВУЗ'].apply(hoice_educaton)
```

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
#ваш код здесь
excepted = [] # ячейки, в которых данные по дате обнаружить не удалось 

def get_experience(text: str) -> int:
    """ Находит опыт работы в месяцах

    Args:
        text (str): Текст для разбора

    Returns:
        int: Опыт работы в месяцах
    """       
    # Если опыт работы не указан, оставляем как есть
    if text is np.nan or text == 'Не указано':
        return np.nan
    
    year = 0
    month = 0
    text = text.lower().replace(',', '').split()
    
    # Находим в тексте число лет
    for i, word in enumerate(text):
        if word in ['год', 'года', 'лет']:
            if text[i-1].isnumeric():
                year = int(text[i-1])
        # Ограничение. Сколько слов проверять.
        if i == 5:
            break
        
    # Находим в тексте число месяцев
    for i, word in enumerate(text):
        if word in ['месяцев', 'месяца', 'месяц']:
            if text[i-1].isnumeric():
                month = int(text[i-1])
        # Ограничение. Сколько слов проверять.
        if i == 7:
            break
    if 12 * year + month == 0:
        excepted.append(text)
    return 12 * year + month


# Создаю новый признак
hh_df['Опыт работы (месяц)'] = hh_df['Опыт работы'].apply(get_experience)
```

```python
# Проверяю корректность полученных данных
print(hh_df['Опыт работы (месяц)'].describe())
print(hh_df['Опыт работы (месяц)'].median())
print(f'В процессе преобразования не полуилось извлеч дату из {len(excepted)} значений')
```

```python
# Удаляю неинформативный признак
hh_df = hh_df.drop(['Опыт работы'], axis=1)
hh_df.duplicated().sum()
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
#ваш код здесь
# Число выводимых на экран исключений
# Носит информативный характер, может не указывать на ошиби
count = 2
# Список городов-миллионников
million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск','Омск', 
                  'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']
# Убираю все заглавные символы в имени городов миллионников
million_cities = [x.lower() for x in million_cities]


def get_city(text: str) -> str:
    """ Классифицирую города на 4 группы:
    Москва, Санкт-Петербург, город-миллионник, другие.

    Args:
        text (str): Название города

    Returns:
        str: один из 4-ех вариантов
    """    
    # Предобработка входных значений
    temp = text.lower().split(',')
    temp = temp[0].split()
    
    if temp[0] == 'москва':
        return 'Москва'
    if temp[0] == 'санкт-петербург':
        return 'Санкт-Петербург'
    if temp[0] in million_cities:
        return 'город-миллионник'
    return 'другие'
        

def ready_to(text: str, substring: str, debug: bool=False) -> bool:
    """Функция вычисляет наличие необходимого признака и мнение пользователя на него

    Args:
        text (string): сырые данные
        substring (string): признак, который мы ищем
        debug (boolean): Выводить ли на экран исключения, при их появлении

    Returns:
        boolean: Готов ли пользователь на это или нет
        
    ПРИМЕР:
        text = 'Готов к переезду'
        
        ready_to(text, 'переез')
        >>> True

        text = 'Не готова к переезду'
        
        ready_to(text, 'переез')
        >>> False  
    """    
    global count
    i_col = 0 # Номер ячейки в списке с данными о перезде
    # Делю данные на части
    temp = text.lower().split(',')
    # В каждой части произвожу поиск
    for i in range(len(temp)):
        if temp[i].find(substring) > -1:
            # Запоминаю номер ячейки
            i_col = i
            break
    if i_col:
        # Если в тексте найен признак, ищу что соискатель о нем думает
        temp = temp[i_col].split()
    
        if temp[0] in ['готов', 'готова', 'хочу']:
            return True
        if temp[0] == 'не':
            return False
    # Если признак не найден, вероятно это ошибка в данных или алгоритме
    count -= 1
    if debug and count >= 0:
        print(text)
        print(f'----------ИСКЛЮЧЕНИЕ {substring}-------------------')
    return False
        
        
# Создаю новые признаки      
hh_df['Город'] = hh_df['Город, переезд, командировки'].apply(get_city)
hh_df['Готовность к переезду'] = hh_df['Город, переезд, командировки'].apply(ready_to, substring=' перее', debug=True)
hh_df['Готовность к командировкам'] = hh_df['Город, переезд, командировки'].apply(ready_to, substring=' командировк', debug=True)
```

```python
# Проверяю корректность полученных данных
print(hh_df['Город'].value_counts()) 
print(round(100 / hh_df.shape[0] * hh_df[hh_df['Город'] == 'Санкт-Петербург'].shape[0],0),'% соисткатлей живут в Санкт-Петербурге.') 
print(hh_df['Готовность к переезду'].value_counts())
print(hh_df['Готовность к командировкам'].value_counts())
mask1 = hh_df['Готовность к переезду'] == True
mask2 = hh_df['Готовность к командировкам'] == True
print(round(100 / hh_df.shape[0] * hh_df[mask1 & mask2].shape[0],0),'% соисткатлей готовы одновремменно к командировкам и переезду.') 
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
#ваш код здесь
def get_employmend(text: str, substring: str) -> bool:
    """ Ищет признак в тексте

    Args:
        text (str): Текст для разбора
        substring (str): Подстрока содержащяя признак или его часть.

    Returns:
        bool: True, если признак найден, иначе False
        
        ПРИМЕР:
        get_employmend('Меня интересует только полная занятость', 'полная занятость') -> True
    """    
    return text.lower().find(substring)>-1


# Создаю новые признаки
hh_df['Полная занятость'] = hh_df['Занятость'].apply(get_employmend, substring='полная занятость')
hh_df['Частичная занятость'] = hh_df['Занятость'].apply(get_employmend, substring='частичная занятость')
hh_df['Проектная работа'] = hh_df['Занятость'].apply(get_employmend, substring='проектная работа')
hh_df['Стажировка'] = hh_df['Занятость'].apply(get_employmend, substring='стажировка')
hh_df['Волонтерство'] = hh_df['Занятость'].apply(get_employmend, substring='волонтерство')
hh_df['Полный день'] = hh_df['График'].apply(get_employmend, substring='полный день')
hh_df['Сменный график'] = hh_df['График'].apply(get_employmend, substring='сменный график')
hh_df['Гибкий график'] = hh_df['График'].apply(get_employmend, substring='гибкий график')
hh_df['Удаленная работа'] = hh_df['График'].apply(get_employmend, substring='удаленная работа')
hh_df['Вахтовый метод'] = hh_df['График'].apply(get_employmend, substring='вахтовый метод')
```

Теперь у нас есть вся необходимая информация для того, чтобы создать признак "ЗП (руб)" - заработная плата в рублях.

После ответа на контрольные вопросы удалите исходный столбец заработной платы "ЗП" и все промежуточные столбцы, если вы их создавали.

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
#ваш код здесь
# Читаю цены закрытия по валютам
exchange = pd.read_csv(r'C:\Users\User\Desktop\p2\git\my_works\data\база\ExchangeRates.csv', sep=',')
# Удаляю неинформативные признаки
exchange.drop(['per','time','vol'], axis=1, inplace=True)
# Формирую недостающие данные
nrows = exchange['date'].nunique()
exchange_rub = pd.DataFrame({'currency': ['РУБ']*nrows,
                             'date': exchange['date'].unique(),
                             'close': [1.]*nrows,
                             'proportion': [1]*nrows,
                             })
# Дополняю загруженные данные недостающими
exchange = pd.concat([exchange, exchange_rub], axis=0, ignore_index=True)
# Преобразую типы и вычисляю недостающие признаки
exchange['date'] = pd.to_datetime(exchange['date'], format='%d/%m/%y').dt.date
exchange['coefficient'] = exchange['close'] / exchange['proportion'] 
hh_df['Дата обновления резюме'] = pd.to_datetime(hh_df['Обновление резюме'], dayfirst=True).dt.date #, dayfirst=True
# hh_df['Дата обновления резюме'] = pd.to_datetime(hh_df['Обновление резюме'], format='%d.%m.%Y %H:%M').dt.date
# Убираю возможные повторы группировкой, оставляю последнее значение
exchange_group = exchange.groupby(by=['currency','date'], as_index=False)['date','coefficient'].last()

def salary_conversion(text: str, num: int):
    """ Функция извлекает информацию о валюте и ЗП

    Args:
        text (str): Текст для разбора
        num (int): Номер слова из текста для анализа

    Returns:
        [str]: Если num == 0, выводит валюту в формате ISO;
        [int64]: Если num == 1, выводит зарплату
    """    
    # Temp получает [ЗП, Валюта, ...]
    temp =text.replace(',', '').replace('.', '').split(' ')
    # Словарь преобразований в станарт ISO
    to_iso = {'ГРН': 'UAH',
              'USD': 'USD',
              'EUR': 'EUR',
              'БЕЛРУБ': 'BYN',
              'KGS': 'KGS',
              'СУМ': 'UZS',
              'AZN': 'AZN',
              'KZT': 'KZT',
              'РУБ': 'РУБ'}
    if num:
        # Возвращает валюту в формате ISO
        return to_iso[temp[1].upper()]
    else:
        # Возварщает ЗП
        return int(temp[0])
    
    
# Вычисляю промежуточные признаки
hh_df['Валюта'] = hh_df['ЗП'].apply(salary_conversion, num=1)
hh_df['ЗП в валюте'] = hh_df['ЗП'].apply(salary_conversion, num=0)
# Дополняю данных с hh коэффициентами для перерасчета ЗП в рубли
hh_df = hh_df.merge(exchange_group,
                    how='left',
                    right_on=('date', 'currency'),
                    left_on=('Дата обновления резюме', 'Валюта')
)
# Проверяю корректность результатов работы алгоритма
print('Число неправильно отработанных полей при объединении', 
      hh_df[hh_df['coefficient'].isna()]['Валюта'].value_counts())
# Расчитываю ЗП в рублях
hh_df['ЗП (руб)'] = hh_df['ЗП в валюте'] * hh_df['coefficient']
```
